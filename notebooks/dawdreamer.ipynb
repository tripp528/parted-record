```python
import dawdreamer as daw
from dawdreamer.faust import (
    createLibContext, destroyLibContext, FaustContext)
from dawdreamer.faust.box import *

import numpy as np
import librosa
from scipy.io import wavfile
from IPython.display import Audio
import IPython.display as ipd

SAMPLE_RATE = 44100
```

```python

def show_audio(data, autoplay=False):
    # Only normalize the audio if it's too loud
    if abs(data).max() > 1.:
        data /= abs(data).max()
    
    ipd.display(Audio(
        data=data, rate=SAMPLE_RATE, normalize=False, autoplay=autoplay
    ))
    
def render(engine, duration=5, display=True):
    engine.render(duration)
    audio = engine.get_audio()
    print(audio.shape)
    if display:
        show_audio(audio)
    return audio
```

```python
engine = daw.RenderEngine(SAMPLE_RATE, 512)  # 512 block size

osc = engine.make_oscillator_processor('osc1', 220)

engine.load_graph([
    (osc, [])
])

render(engine)
```

```python
engine = daw.RenderEngine(SAMPLE_RATE, 512)  # 512 block size

faust_processor.set_dsp_string(
    """
    declare name "MySine";
    freq = hslider("freq", 220, 0, 20000, 0);
    gain = hslider("vol[unit:dB]", 0, -120, 20, 0) : ba.db2linear;
    process = freq : os.osc : _*gain <: si.bus(2);
    """
    )

engine.load_graph([
    (faust_processor, [])
])
engine.render(4., beats=True)  # render 4 beats.
audio = engine.get_audio()  # shaped (2, N samples)
show_audio(audio)
```

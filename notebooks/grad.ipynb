```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = torch.tensor([0.], requires_grad=True)
b = torch.tensor([0.], requires_grad=True)

def f_x(x):
    return m * (x/2)**2 + b

def target_fn(x): 
    return 5 * (x/2)**2 + 3

def fmt(x): 
    return f'{np.round(np.asarray(x), 2) : .3f}'
    
lr = 0.01
for x_in in [1] * 1000:
    result = f_x(x_in)
    target = target_fn(x_in)
    loss = (result - target) ** 2
    loss.backward()
    with torch.no_grad():
        m -= m.grad * lr
        b -= b.grad * lr
    print((
        # f'm: {rnd(m.item())} \t'
        # f'b: {rnd(b.item())} \t'
        # f't: {fmt(target)} \t'
        # f'r: {fmt(result.item())} \t'
        f'l: {fmt(loss.item())} \t'
        f'm_grad: {fmt(m.grad.item())} \t'
        f'b_grad: {fmt(b.grad.item())} \t'
        
    ))
    m.grad.zero_()
    b.grad.zero_()
```
